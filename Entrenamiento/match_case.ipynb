```python
# Importando la Libreria/biblioteca de matemáticas de Python
import math

# DEFINICION DE LAS FUNCIONES:
def mostrar_título():
    print('Ejemplo Calculadora Sencilla (+,-,*,/):')

def mostrar_menu():
    print()
    print("MENU:")
    print("0: Salir")
    print("1: Sumar (+)")
    print("2: Restar (-)")
    print("3: Multiplicar (*)")
    print("4: Dividir (/)")
    print("5: Logaritmo (log10)")
    print("6: Coseno (cos)")
    print("7: Tangente (tan)")
    print("8: Seno (sen)")
    print("9: potenciacion (x^y)")
    print("10: raiz cuadrada")
    print()

def ingresar_datos():
    opcion = input("Ingrese la opción deseada del menú: ")

    # si elige salir, no pido números
    if opcion == "0":
        return opcion, None, None
    # opciones que requieren un solo número
    if opcion in ["5", "6", "7", "8", "10"]:
        try:
            numero_uno = float(input("Ingrese el número para la operación: "))
            return opcion, numero_uno, None
        except ValueError:
            print("Entrada inválida. Por favor ingrese un número.")
            return "invalid", None, None
    # opciones que requieren dos números
    if opcion in ["1", "2", "3", "4", "9"]:
        try:
            numero_uno = float(input("Ingrese el primer número de la operación: "))
            numero_dos = float(input("Ingrese el segundo número de la operación: "))
            return opcion, numero_uno, numero_dos
        except ValueError:
            print("Entrada inválida. Por favor ingrese números.")
            return "invalid", None, None
    # opción no válida
    else:
        print("Opción no válida.")
        return "invalid", None, None


# FUNCIONES DE OPERACIONES
def calcular_suma(a, b): return a + b
def calcular_resta(a, b): return a - b
def calcular_multiplicacion(a, b): return a * b
def calcular_division(a, b): return a / b
def calcular_logaritmo_base10(x): return math.log10(x)
def calcular_coseno(x): return math.cos(x)
def calcular_tangente(x): return math.tan(x)
def calcular_seno(x): return math.sin(x)
def calcular_potenciacion(x, y): return x**y
def calcular_raiz_cuadrada(x): return math.sqrt(x)

def procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos):
    match opcion:
        case "1":
            print("Resultado:", calcular_suma(numero_uno, numero_dos))
        case "2":
            print("Resultado:", calcular_resta(numero_uno, numero_dos))
        case "3":
            print("Resultado:", calcular_multiplicacion(numero_uno, numero_dos))
        case "4":
            if numero_dos != 0:
                print("Resultado:", calcular_division(numero_uno, numero_dos))
            else:
                print("¡No es posible dividir entre cero!")
        case "5":
            if numero_uno is not None:
              print("Resultado:", calcular_logaritmo_base10(numero_uno))
            else:
              print("Ingrese un número válido para logaritmo.")
        case "6":
            if numero_uno is not None:
              print("Resultado:", calcular_coseno(numero_uno))
            else:
              print("Ingrese un número válido para coseno.")
        case "7":
            if numero_uno is not None:
              print("Resultado:", calcular_tangente(numero_uno))
            else:
              print("Ingrese un número válido para tangente.")
        case "8":
            if numero_uno is not None:
              print("Resultado:", calcular_seno(numero_uno))
            else:
              print("Ingrese un número válido para seno.")
        case "9":
            print("Resultado:", calcular_potenciacion(numero_uno, numero_dos))
        case "10":
            if numero_uno is not None:
              print("Resultado:", calcular_raiz_cuadrada(numero_uno))
            else:
              print("Ingrese un número válido para raíz cuadrada.")
        case _:
            print("Opción no válida.")


def llamar_funciones():
    mostrar_título()
    while True:  # bucle principal
        mostrar_menu()
        opcion, numero_uno, numero_dos = ingresar_datos()

        if opcion == "0":
            print("¡Hasta luego!")
            break  # salimos del bucle y termina el programa
        if opcion == "invalid":
          continue # continue the loop if the input is invalid


        procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos)

# EJECUCION DEL PROGRAMA:
llamar_funciones()
```
