```python
# Esta es una calculadora con declaracion de funciones
import math

#DEFINICION DE LAS FUNCIONES:
def mostrar_título():
  '''
  input:no hay entradas
  process: mostrar titulo (está implicito en el nombre de la función, se sugiere que sea un verbo)
  output: no hay salidas
  '''


def mostrar_menu():
  '''input: no recibe nada (0 entradas)
  process:mostrar menu
  output: no retorna nada (0 salidas)'''
  print()
  print("MENU:")
  print("1: Sumar (+)")
  print("2: Restar (-)")
  print("3: Multiplicar (*)")
  print("4: Dividir (/)")
  print("5: Logaritmo (log10)")
  print("6: Coseno (cos)")
  print("7: Tangente (tan)")
  print("8: Seno (sen)")
  print("9: potenciacion (x^y)")
  print("10: raiz cuadrada")
  print()

def ingresar_datos():
  '''input: no hay entradas
  process:ingresar datos (recibir/leer del usuario)
  output: tres variables que contienen localmente las decisiones del usuario (opción, numero_uno, numero_dos)'''
  opcion= input("Ingrese la opción deseada del menú:")
  numero_uno= float(input("Ingrese el primer número de la operación:")) #Guarda como flotante (número real)
  numero_dos= float(input("Ingrese el segundo número de la operación:")) #Guarda como flotante (número real)
  return opcion, numero_uno, numero_dos

#POR FACILIDAD, LAS 7 FUNCIONES SIGUIENTES TIENEN ENTRADAS Y SALIDAS

def calcular_suma(numero_uno, numero_dos):
  '''input: 2 numeros
  process: calcular suma (sumar)
  output: suma'''
  suma=numero_uno+numero_dos
  return suma
def calcular_resta(numero_uno, numero_dos):
  '''input: 2 numeros
  process: calcular resta
  output: resta'''
  resta = numero_uno - numero_dos
  return resta

def calcular_multiplicacion(numero_uno, numero_dos):
  '''input: 2 numeros
  process: calcular multiplicacion
  output: multiplicacion'''
  multiplicacion = numero_uno * numero_dos
  return multiplicacion

def calcular_division(numero_uno, numero_dos):
  '''input: 2 numeros
  process: calcular division
  output: division'''
  if numero_dos != 0:
    division = numero_uno / numero_dos
    return division
  else:
    print("Error: No se puede dividir entre cero.")
    return None

def calcular_logaritmo(numero_uno):
  '''input: 1 numero
  process: calcular logaritmo en base 10
  output: logaritmo'''
  if numero_uno > 0:
    logaritmo = math.log10(numero_uno)
    return logaritmo
  else:
    print("Error: El logaritmo solo se puede calcular para números positivos.")
    return None

def calcular_coseno(numero_uno):
  '''input: 1 numero
  process: calcular coseno
  output: coseno'''
  coseno = math.cos(numero_uno)
  return coseno

def calcular_tangente(numero_uno):
  '''input: 1 numero
  process: calcular tangente
  output: tangente'''
  tangente = math.tan(numero_uno)
  return tangente

def calcular_seno(numero_uno):
  '''input: 1 numero
  process: calcular seno
  output: seno'''
  seno = math.sin(numero_uno)
  return seno

def calcular_potencia(base, exponente):
  '''input: 2 numeros
  process: calcular potencia
  output: potencia'''
  potencia = base ** exponente
  return potencia

def calcular_raiz_cuadrada(numero):
  '''input: 1 numero
  process: calcular raiz cuadrada
  output: raiz cuadrada'''
  if numero >= 0:
    raiz = math.sqrt(numero)
    return raiz
  else:
    print("Error: No se puede calcular la raíz cuadrada de un número negativo.")
    return None

def main():
  mostrar_título()
  mostrar_menu()
  opcion, numero_uno, numero_dos = ingresar_datos()

  if opcion == '1':
    resultado = calcular_suma(numero_uno, numero_dos)
  elif opcion == '2':
    resultado = calcular_resta(numero_uno, numero_dos)
  elif opcion == '3':
    resultado = calcular_multiplicacion(numero_uno, numero_dos)
  elif opcion == '4':
    resultado = calcular_division(numero_uno, numero_dos)
  elif opcion == '5':
    resultado = calcular_logaritmo(numero_uno)
  elif opcion == '6':
    resultado = calcular_coseno(numero_uno)
  elif opcion == '7':
    resultado = calcular_tangente(numero_uno)
  elif opcion == '8':
    resultado = calcular_seno(numero_uno)
  elif opcion == '9':
    resultado = calcular_potencia(numero_uno, numero_dos)
  elif opcion == '10':
    resultado = calcular_raiz_cuadrada(numero_uno)
  else:
    print("Opción no válida. Por favor, seleccione una opción del 1 al 10.")
    resultado = None

  if resultado is not None:
    print("El resultado de la operación es:", resultado)

if __name__ == "__main__":
  main()
```
